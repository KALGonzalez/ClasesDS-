# ***Gráficos de disperción. ***

Un gráfico de dispersión es útil para mostrar la relación entre dos variables. Cada punto en el gráfico representa un par de valores (X, Y). Aquí tienes un desglose simplificado de sus usos y ejemplos:

Usos del Gráfico de Dispersión
Analizar correlación: Para ver si hay una relación entre dos variables.
Ejemplo: Altura y peso de personas.
Identificar patrones: Para detectar agrupaciones o tendencias.
Ejemplo: Ventas de productos en relación al precio.
Detección de valores atípicos: Para ver puntos que se desvían mucho del patrón.
Ejemplo: Identificar ventas muy altas o bajas en una tienda en días específicos.

```python
import matplotlib.pyplot as plt

# Variables (X: horas de estudio, Y: calificaciones)
horas_estudio = [1, 2, 3, 4, 5, 6, 7]
calificaciones = [60, 65, 70, 72, 80, 85, 90]

# Crear gráfico de dispersión
plt.scatter(horas_estudio, calificaciones)
plt.title("Relación entre horas de estudio y calificaciones")
plt.xlabel("Horas de estudio")
plt.ylabel("Calificaciones")

# Mostrar el gráfico
plt.show()
```

**Diagrama de Cajas y Bigotes**

Un diagrama de cajas y bigotes (o boxplot) es una herramienta visual que muestra la distribución de un conjunto de datos. Es útil para identificar valores atípicos, medir la dispersión y visualizar la mediana y los cuartiles de los datos. Aquí te doy algunos ejemplos comunes de cuándo y cómo se utiliza.

Ejemplos de Uso del Diagrama de Cajas y Bigotes
Comparación de salarios en diferentes departamentos: Puedes ver la dispersión de salarios entre varios departamentos para identificar dónde están los salarios más altos y detectar posibles diferencias.

Distribución de calificaciones de estudiantes: Al usar un diagrama de cajas para las calificaciones en diferentes materias, puedes ver rápidamente qué materia tiene más variabilidad en las calificaciones y si hay valores atípicos (por ejemplo, estudiantes que sacaron notas muy altas o bajas).

Rendimiento en distintos meses o años: En una empresa, puedes analizar los ingresos por mes o año. Esto te permite identificar variaciones, ver meses con ingresos consistentemente altos o bajos y detectar posibles valores atípicos.

Datos de temperatura: Con un diagrama de cajas puedes ver la variación de temperatura en diferentes estaciones o ciudades. Esto ayuda a comparar rangos de temperatura y ver si algunas estaciones tienen un clima más extremo.

Ejemplo de Código Simplificado en Python
Aquí tienes un ejemplo de un diagrama de cajas y bigotes usando Python y Matplotlib.

```python
import matplotlib.pyplot as plt

# Datos: Calificaciones de estudiantes en tres materias
calificaciones_matematicas = [78, 85, 88, 92, 95, 85, 70, 65, 80, 90]
calificaciones_espanol = [82, 79, 85, 88, 90, 85, 89, 76, 80, 84]
calificaciones_ciencias = [75, 78, 80, 85, 88, 90, 72, 85, 95, 85]

# Crear el diagrama de cajas y bigotes
datos = [calificaciones_matematicas, calificaciones_espanol, calificaciones_ciencias]
plt.boxplot(datos, labels=["Matemáticas", "Español", "Ciencias"])

# Etiquetas y título
plt.title("Distribución de Calificaciones por Materia")
plt.ylabel("Calificaciones")

# Mostrar el gráfico
plt.show()
```

**Función Scatter plot**
En Python, puedes crear un scatter plot (gráfico de dispersión) usando matplotlib o seaborn, que son bibliotecas de visualización de datos populares. Aquí te muestro cómo hacerlo con ambas.

Ejemplo 1: Crear un Scatter Plot con matplotlib

**Ejemplo 2 **Scatter Plot con Color y Tamaño Basados en Otras Variables
Si tienes variables adicionales, puedes usarlas para cambiar el color y el tamaño de los puntos.

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78, 77, 85, 86]
sizes = [20, 50, 100, 200, 500, 100, 300, 400, 60, 90, 100, 10, 300]
colors = [5, 15, 25, 35, 45, 55, 65, 75, 85, 95, 100, 35, 45]

# Crear el scatter plot
plt.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=0.5, edgecolor='k')
plt.colorbar(label='Valor de Color')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.title('Scatter Plot con Variaciones de Tamaño y Color')

# Mostrar el gráfico
plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Datos de ejemplo
data = {
    'x': [5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6],
    'y': [99, 86, 87, 88, 100, 86, 103, 87, 94, 78, 77, 85, 86],
    'sizes': [20, 50, 100, 200, 500, 100, 300, 400, 60, 90, 100, 10, 300],
    'colors': [5, 15, 25, 35, 45, 55, 65, 75, 85, 95, 100, 35, 45]
}
df = pd.DataFrame(data)

# Crear el scatter plot
sns.scatterplot(data=df, x='x', y='y', size='sizes', hue='colors', palette='viridis', legend=True)
plt.title('Scatter Plot con Seaborn')

# Mostrar el gráfico
plt.show()
```

**ejemplo**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios con una distribución normal
data = np.random.randn(1000)  # 1000 datos con media 0 y desviación estándar 1

# Crear el gráfico de histograma
plt.figure(figsize=(10, 5))  # Tamaño del gráfico
plt.hist(data, bins=30, alpha=0.7, color='skyblue', edgecolor='black')  # Histograma con 30 intervalos (bins)

# Títulos y etiquetas
plt.title("Histograma de datos aleatorios")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")

# Mostrar el gráfico
plt.show()
```
