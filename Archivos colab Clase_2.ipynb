Concatenar:
1. Concatenar Strings:
Con el operador +:
python
Copy code
str1 = "Hello"
str2 = "World"
result = str1 + " " + str2  # Output: "Hello World"


Con join (útil para listas de strings):
python
Copy code
words = ["Hello", "World"]
result = " ".join(words)  # Output: "Hello World"


2. Concatenar Listas:
Con el operador +:
python
Copy code
list1 = [1, 2, 3]
list2 = [4, 5, 6]
result = list1 + list2  # Output: [1, 2, 3, 4, 5, 6]


Con extend (modifica la lista en lugar de crear una nueva):
python
Copy code
list1 = [1, 2, 3]
list2 = [4, 5, 6]
list1.extend(list2)  # list1 ahora es [1, 2, 3, 4, 5, 6]


3. Concatenar Arrays o Matrices en NumPy:
Usando np.concatenate():
python
Copy code
import numpy as np
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
result = np.concatenate((arr1, arr2))  # Output: [1 2 3 4 5 6]


También puedes especificar el eje en matrices multidimensionales con el argumento axis.
4. Concatenar DataFrames en Pandas:
Con pd.concat():
python
Copy code
import pandas as pd
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
result = pd.concat([df1, df2], ignore_index=True)


O usando pd.merge() si necesitas una combinación más específica de columnas basadas en claves comunes.
